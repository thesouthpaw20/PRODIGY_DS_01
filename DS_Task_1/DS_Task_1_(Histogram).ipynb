```python

# Load the dataset
file_path ="C:/Users/jobkg/OneDrive/Desktop/Internship/Dataset/API_SP.POP.TOTL_DS2_en_csv_v2_739890.csv"
data= pd.read_csv(file_path, skiprows=4)

# Display the first few rows of the dataset
data.head()
```

```python
# Drop non-numeric columns
numeric_data = data.drop(columns=["Country Name", "Country Code", "Indicator Name", "Indicator Code", "Unnamed: 68"])
```

```python


# 1. Population distribution for the year 2020
year = '2020'
population_2020 = data[year].dropna()

plt.figure(figsize=(10, 6))
plt.hist(population_2020, bins=30, edgecolor='black')
plt.title('Population Distribution for the Year 2020')
plt.xlabel('Population')
plt.ylabel('Number of Countries')
plt.yscale('log')
plt.grid(True)
plt.show()
```

```python
# 2. Population distribution for a selected year range (2000-2020) for India
country = 'India'
years = [str(year) for year in range(2000, 2021)]
population_india = data[data['Country Name'] == country][years].values.flatten()

plt.figure(figsize=(10, 6))
plt.hist(population_india, bins=21, edgecolor='black')
plt.title('Population Distribution in India (2000-2020)')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python

# Population Distribution in 2023
population_2023 = numeric_data["2023"].dropna()

# Population Distribution for 2000
population_2000 = numeric_data["2000"].dropna()

# Calculate population growth from 1960 to 2023
population_growth = (numeric_data["2023"] - numeric_data["1960"]).dropna()

# Plotting histograms
plt.figure(figsize=(18, 5))

# Histogram for Population Distribution in 2023
plt.subplot(1, 3, 1)
plt.hist(population_2023, bins=30, color='skyblue', edgecolor='black')
plt.title('Population Distribution in 2023')
plt.xlabel('Population')
plt.ylabel('No. of Countries')

# Histogram for Population Growth from 1960 to 2023
plt.subplot(1, 3, 2)
plt.hist(population_growth, bins=30, color='lightgreen', edgecolor='black')
plt.title('Population Growth from 1960 to 2023')
plt.xlabel('Population Growth')
plt.ylabel('No. of Countries')

# Histogram for Population Distribution in 2000
plt.subplot(1, 3, 3)
plt.hist(population_2000, bins=30, color='lightcoral', edgecolor='black')
plt.title('Population Distribution in 2000')
plt.xlabel('Population')
plt.ylabel('No. of Countries')

plt.tight_layout()
plt.show()
```
